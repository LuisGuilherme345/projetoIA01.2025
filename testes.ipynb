```python
import pandas as pd

df = pd.read_csv("estacionamento.csv")
df
```

```python
df.plot(kind='scatter', x="tempominu", y='preco');
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["tempominu"]]
y = df[["preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[190]])[0][0])
```
